```python
def Date_Conv(X):
    if(int(X['DATE'][-2:]) >= 20):
        return('19' + X['DATE'][-2:])
    else:
        return('20' + X['DATE'][-2:])
```

```python
HH_Income['DATE'] = HH_Income.apply(Date_Conv, axis = 1)
```

```python
Median_CA = HH_Income['Median_CA'].tolist()
Median_CO = HH_Income['Median_CO'].tolist()
Median_NY = HH_Income['Median_NY'].tolist()
Median_TX = HH_Income['Median_TX'].tolist()
Median_IN = HH_Income['Median_IN'].tolist()
```

```python
def Grt(Income):
    ReturnVal = 0
    ReturnIndex = -1
    for i in range(len(Income) - 1):
        Base = Income[i + 1] - Income[i]
        if(Base >= ReturnVal):
            Return_Index = i
            ReturnVal = Base
    return(ReturnVal, Return_Index)
```

```python
City_Greatest_Income = {'Median_CA': Grt(Median_CA), 
                       'Median_CO': Grt(Median_CO), 
                        'Median_NY': Grt(Median_NY),
                       'Median_TX': Grt(Median_TX), 
                        'Median_IN': Grt(Median_IN)}
```

```python
City_Greatest_Income
```

```python
 Cty_Index = max(City_Greatest_Income.items(), key=lambda k: k[1])
```

```python
Cty_Index
```

```python
print(HH_Income.at[Cty_Index[1][1], 'DATE'])
```
